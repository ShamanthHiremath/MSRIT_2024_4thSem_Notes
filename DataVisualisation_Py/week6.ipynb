```python
'''
. - Any Character Except New Line
\d - Digit (0-9)
\D - Not a Digit (0-9)
\w - Word Character (a-z, A-Z, 0-9, _)
\W - Not a Word Character
\s - Whitespace (space, tab, newline)
\S - Not Whitespace (space, tab, newline)
\b - Word Boundary
\B - Not a Word Boundary
^ - Beginning of a String
$ - End of a String
[] - Matches Characters in brackets
[^ ] - Matches Characters NOT in brackets
| - Either Or
( ) - Group
Quantifiers:
* - 0 or More
+ - 1 or More
? - 0 or One
{3} - Exact Number
{3,4} - Range of Numbers (Minimum, Maximum)

'''
```

```python
strr = input("Enter the text: ")
vow = re.findall("[aeiou]", strr)
digits = re.findall("[0-9]", strr)
cons = re.findall("[^aeiou]", strr)

print(f"Vowels : {vow}")
print(f"Consonants : {cons}")
print(f"Digits : {digits}")
```

```python
'''
Develop a python program to find all three, four and five characters long word in a
string.

'''
```

```python
input_string = "bfjasbdjab ddbsajfbaj  asdbdj  asdbj  hdbja  abj  jab iaibj  adhjad"

three_char_words = re.findall(r'\b\w{3}\b', input_string)
four_char_words = re.findall(r'\b\w{4}\b', input_string)
five_char_words = re.findall(r'\b\w{5}\b', input_string)
print("3 words: ", three_char_words)
print("4 words: ", four_char_words)
print("5 words: ", five_char_words)

```
