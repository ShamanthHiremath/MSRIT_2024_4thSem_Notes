```python
import random

class ForthSem:
    def __init__(self):
        self.RollNumber = [i for i in range(20)]
        self.Test1Marks = [random.randint(50, 100) for i in range(20)]
        self.Test2Marks = [random.randint(50, 100) for i in range(20)]
        self.Test3Marks = [random.randint(50, 100) for i in range(20)]
        self.avg = [-1 for i in range(20)]
        self.dic = {}
        
    def studentAVG(self):
        for i in range(20):
            self.avg[i] = (self.Test1Marks[i] + self.Test2Marks[i] + self.Test3Marks[i])/3
            self.dic[self.RollNumber[i]] = self.avg[i]
        for i in self.dic.items():
            print(i)
    
    def AVG(self):
        print(f"Test 1 class avg : {sum(self.Test1Marks)/20}")
        print(f"Test 2 class avg : {sum(self.Test2Marks)/20}")
        print(f"Test 3 class avg : {sum(self.Test3Marks)/20}\n")
        
testing = ForthSem()
testing.AVG()
testing.studentAVG()
        
```

```python
class TimeClass:
    def __init__(self, sec, mins, hrs):
        self.sec = sec
        self.mins = mins
        self.hrs = hrs
        if self.sec >= 60:
            self.mins += self.sec // 60
            self.sec = self.sec % 60
        if self.mins >= 60:
            self.hrs += self.mins // 60
            self.mins = self.mins % 60
        
    def displayTime(self):
        print(f"Time: {self.hrs}:{self.mins}:{self.sec}")
        
    def addTime(self, time):
        self.sec += time.sec
        self.mins += time.mins
        self.hrs += time.hrs
        if self.sec >= 60:
            self.mins += self.sec // 60
            self.sec = self.sec % 60
        if self.mins >= 60:
            self.hrs += self.mins // 60
            self.mins = self.mins % 60
        self.displayTime()
        
    def subTime(self, time):
        self.sec -= time.sec
        self.mins -= time.mins
        self.hrs -= time.hrs
        if self.sec < 0:
            self.sec += 60
            self.mins -= 1
        if self.mins < 0:
            self.mins += 60
            self.hrs -= 1
        self.displayTime()
        
    def countSeconds(self):
        totalSec = self.sec + self.mins * 60 + self.hrs * 60 * 60
        print(f"Total Seconds: {totalSec}")

time1 = TimeClass(30, 6, 2)
time1.displayTime()
time2 = TimeClass(45, 12, 23)
time2.addTime(time1)
time2.subTime(time1)
time1.countSeconds()
```

```python
import math

class Shape:
    def __init__(self, side1, side2):
        self.side1 = side1
        self.side2 = side2
    
    def calculateArea(self):
        pass
    
    def displayDetails(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, width):
        super().__init__(length, width)
    
    def calculateArea(self):
        return self.side1 * self.side2
    
    def displayDetails(self):
        print("Rectangle Details:")
        print("Length:", self.side1)
        print("Width:", self.side2)
        print("Area:", self.calculateArea())

class Triangle(Shape):
    def __init__(self, base, height):
        super().__init__(base, height)
    
    def calculateArea(self):
        return 0.5 * self.side1 * self.side2
    
    def displayDetails(self):
        print("Triangle Details:")
        print("Base:", self.side1)
        print("Height:", self.side2)
        print("Area:", self.calculateArea())

rect = Rectangle(5, 4)
rect.displayDetails()

print()

tri = Triangle(6, 8)
tri.displayDetails()
```
