```python
class Stack:
    arr =[]
    
    def isEmpty(self):
        return len(self.arr) == 0

    def push(self, value):
        self.arr.append(value)
        return 

    def pop(self):
        if isEmpty(self.arr):
            print("Stack empty")
            return -1
        return self.arr.pop()

    def peek(self):
        if self.isEmpty():
            print("Stack empty")
            return -1
        return self.arr[-1]

class Queue :
    arr = []
    
    def isEmpty(self):
        return len(self.arr) == 0
    
    def enqueue(self, value):
        self.arr.append(value)
    
    def dequeue(self):
        if self.isEmpty():
            return -1
        res = self.arr[0]
        self.arr = self.arr[1:]
        return res
    
    def peek(self):
        if self.isEmpty():
            return -1
        return self.arr[0]


s =  Stack()
q = Queue()
s.push(5)
q.enqueue(7)
s.push(6)
q.enqueue(8)
print(s.peek())
print(q.peek())
print(s.pop())
print(q.dequeue())
```
